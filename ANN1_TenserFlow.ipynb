```python
# !pip install tensorflow
# !pip install keras
import pandas as pd
import numpy as np
import warnings  # >>> 3shna lw 3awz ashyl warning zahr fe cell wla haga
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv(r"/content/HousingData.csv")
df.head(5)
```

```python

warnings.filterwarnings("ignore")

x = df.drop(["MEDV"],1)
y = df["MEDV"]
```

```python
null_columns = x.columns[x.isnull().any()]
null_rows = x[x.isnull().any(axis=1)]

x = x.dropna()
y = y.drop(null_rows.index)  # Remove corresponding target values

x = x.reset_index(drop=True)
y = y.reset_index(drop=True)
```

```python
scaler = StandardScaler()
columns = x.columns    # >> 3shan lma b3ml fit transform byzharly arrays w msh bkon 3arf esm elcolumns
x = scaler.fit_transform(x)
x = pd.DataFrame(x,columns = columns)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3)
print("x_train shape ",x_train.shape)
print("y_train shape ",y_train.shape)
print("x_test shape ",x_test.shape)
print("y_tes t shape ",y_test.shape)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers  # Import the 'layers' module
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(layers.Dense(32,activation='sigmoid',input_shape=(x_train.shape[1],)))
model.add(layers.Dense(1,activation="linear"))
model.summary()
```

```python
model.compile(optimizer='sgd',loss='mean_squared_error')
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=20)
```

```python
preds = model.predict(x_test).reshape(-1,)
res = pd.DataFrame()
res["Actual"] = y_test.values
res["preds"] = preds
res
```
